```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from math import sin, pi, cos
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
milk=pd.read_csv('../Files/milk.csv') 
```

```python
milk.head(10)
```

```python
plt.plot(milk['Month'], milk['Production'],  '-o')
```

```python

month=np.array(milk['Month']).reshape(1,-1)
production=np.array(milk['Production']).reshape(1,-1)

model = make_pipeline(PolynomialFeatures(2), LinearRegression())
model = model.fit(month, production + np.sin(2*pi*month/12) + np.cos(2*pi*month/12)) 
```

```python
#x_pred=np.array(range(1, 100)).reshape(1,-1)
x_pred=np.array(range(1, 169)).reshape(1,-1)
```

```python
y_pred=model.predict(x_pred)
```

```python
plt.scatter(milk['Month'], milk['Production'])
plt.plot(x_pred[0], y_pred[0], color='red')
plt.show()
```

```python

r2_score(production[0], y_pred[0])
```
