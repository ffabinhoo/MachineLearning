```python

import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.linspace(0, 1, 1000) #time interval

#build signals that I will sum up
A = np.sin(2*np.pi*t*5)# AMPLITUDE 5
B = np.sin(2*np.pi*t*7)# AMPLITUDE 7
C = np.sin(2*np.pi*t*9)# AMPLITUDE 9

soma = 3*A + 4*B + 5*C # note that I am multipling by their amplitudes 3,4 e 5
```

```python
plt.rcParams['figure.figsize'] = (15,3)

plt.subplot(1,4,1)
plt.plot(t, A)
plt.title("Frequency 5")
plt.grid()

plt.subplot(1,4,2)
plt.plot(t, B)
plt.title("Frequency 7")
plt.grid()

plt.subplot(1,4,3)
plt.plot(t, C)
plt.title("Frequency 9")
plt.grid()

plt.subplot(1,4,4)
plt.plot(t, soma,'r')
plt.title("Signal summed up")
plt.grid()
```

## Frequency Domain in Python


**Fourier transformation Equation in python**
\begin{equation}
X[k] = \sum_{n = 0}^{N-1} x[n] \cdot e^{-\frac{2\pi i}{N}}
\end{equation}

```python
#take it to the frequency domain
plt.rcParams['figure.figsize'] = (5,3)

N = len(t)# Esse N  importante pois verifica a amplitude do sinal logo abaixo
D = 2*(1/N)*np.fft.fft(soma) # usei a transformada de Fourier aqui
plt.plot(np.abs(D[0:20]))# dividi por N/2 para mostrar a amplitude correta, 'abs' porque quero nmeros absolutos positivos e coloquei at amostra 20, mais voc pode mudar
plt.grid()
#note no grfico as freqncias (5 com amplitude 3, 7 com amplitude 4, 9 com amplitude 5) voc pode soamr os sinais de novo pra ver se d o original
```

```python
#Frequqncia de varredura de 0  tal ponto neste caso 250
#Neste caso voc vai ver todas as frequ~encias plotadas

freq_inicial = 0 
freq_final =   250
Y = np.array(D)

N = len(t)
for f in range(freq_inicial, freq_final):  
  c = (1/N) * Y[f] * np.exp(2j*np.pi*f*t)
  plt.plot(c.real)
plt.grid()
```

```python
# PASSANDO PARA O DOMNIO TEMPO

A = np.fft.fft(soma)
B = np.real(np.fft.ifft(A)) # Transformada Inversa de Fourier
plt.plot(B)
plt.grid()
```
