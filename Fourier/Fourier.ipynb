```
#importação bibliotecas
import numpy as np
from matplotlib import pyplot as plt
```

```
t = np.linspace(0, 1, 1000) #intervalo de Tempo

#Construção dos 3 sinais que eu vou somar
A = np.sin(2*np.pi*t*5)# AMPLITUDE 5
B = np.sin(2*np.pi*t*7)# AMPLITUDE 7
C = np.sin(2*np.pi*t*9)# AMPLITUDE 9

soma = 3*A + 4*B + 5*C # note que estou multiplicando pelas amplitudes 3,4 e 5
```

```
plt.rcParams['figure.figsize'] = (15,3)

plt.subplot(1,4,1)
plt.plot(t, A)
plt.title("Frequência 5")
plt.grid()

plt.subplot(1,4,2)
plt.plot(t, B)
plt.title("Frequência 7")
plt.grid()

plt.subplot(1,4,3)
plt.plot(t, C)
plt.title("Frequência 9")
plt.grid()

plt.subplot(1,4,4)
plt.plot(t, soma,'r')
plt.title("Sinal Somado")
plt.grid()
```

## Domínio da Frequência Python


**Equação da Transformada de Fourier**
\begin{equation}
X[k] = \sum_{n = 0}^{N-1} x[n] \cdot e^{-\frac{2\pi i}{N}}
\end{equation}

```
#levar para domínio da frequência
plt.rcParams['figure.figsize'] = (5,3)

N = len(t)# Esse N é importante pois verifica a amplitude do sinal logo abaixo
D = 2*(1/N)*np.fft.fft(soma) # usei a transformada de Fourier aqui
plt.plot(np.abs(D[0:20]))# dividi por N/2 para mostrar a amplitude correta, 'abs' porque quero números absolutos positivos e coloquei até amostra 20, mais você pode mudar
plt.grid()
#note no gráfico as freqências (5 com amplitude 3, 7 com amplitude 4, 9 com amplitude 5) você pode soamr os sinais de novo pra ver se dá o original
```

```
#Frequêqncia de varredura de 0 à tal ponto neste caso 250
#Neste caso você vai ver todas as frequ~encias plotadas

freq_inicial = 0 
freq_final =   250
Y = np.array(D)

N = len(t)
for f in range(freq_inicial, freq_final):  
  c = (1/N) * Y[f] * np.exp(2j*np.pi*f*t)
  plt.plot(c.real)
plt.grid()
```

```
# PASSANDO PARA O DOMÍNIO TEMPO

A = np.fft.fft(soma)
B = np.real(np.fft.ifft(A)) # Transformada Inversa de Fourier
plt.plot(B)
plt.grid()
```
